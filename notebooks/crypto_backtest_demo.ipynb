## 5. Compare All Assets Side-by-Side

Compare BTC, SPDR, and GLAM performance.

```python
# Create comparison DataFrame
comparison = pd.DataFrame({
    'Asset': ['BTC', 'SPDR', 'GLAM'],
    'Return %': [metrics_btc['total_return'], metrics_spdr['total_return'], metrics_glam['total_return']],
    'Win Rate %': [metrics_btc['win_rate'], metrics_spdr['win_rate'], metrics_glam['win_rate']],
    'Sharpe': [metrics_btc['sharpe_ratio'], metrics_spdr['sharpe_ratio'], metrics_glam['sharpe_ratio']],
    'Max DD %': [metrics_btc['max_drawdown'], metrics_spdr['max_drawdown'], metrics_glam['max_drawdown']],
    'Trades': [metrics_btc['total_trades'], metrics_spdr['total_trades'], metrics_glam['total_trades']]
})

print("\n" + "="*70)
print("ASSET COMPARISON - SMA Crossover Strategy")
print("="*70)
display(comparison)

# Find best performer
best_idx = comparison['Return %'].idxmax()
print(f"\nðŸ† Best Performer: {comparison.loc[best_idx, 'Asset']} with {comparison.loc[best_idx, 'Return %']:.2f}% return")
```

```python
# Visualize comparison
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Return comparison
ax1 = axes[0, 0]
colors = ['green' if x > 0 else 'red' for x in comparison['Return %']]
ax1.bar(comparison['Asset'], comparison['Return %'], color=colors, alpha=0.7)
ax1.set_title('Total Return Comparison', fontweight='bold')
ax1.set_ylabel('Return (%)')
ax1.grid(True, alpha=0.3, axis='y')

# Win Rate comparison
ax2 = axes[0, 1]
ax2.bar(comparison['Asset'], comparison['Win Rate %'], color='skyblue', alpha=0.7)
ax2.set_title('Win Rate Comparison', fontweight='bold')
ax2.set_ylabel('Win Rate (%)')
ax2.grid(True, alpha=0.3, axis='y')

# Sharpe comparison
ax3 = axes[1, 0]
ax3.bar(comparison['Asset'], comparison['Sharpe'], color='orange', alpha=0.7)
ax3.set_title('Sharpe Ratio Comparison', fontweight='bold')
ax3.set_ylabel('Sharpe Ratio')
ax3.grid(True, alpha=0.3, axis='y')

# Max Drawdown comparison
ax4 = axes[1, 1]
ax4.bar(comparison['Asset'], comparison['Max DD %'], color='coral', alpha=0.7)
ax4.set_title('Max Drawdown Comparison', fontweight='bold')
ax4.set_ylabel('Max Drawdown (%)')
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

## 6. Test All 5 Strategies on One Asset

Compare different strategies on SPDR.

```python
# Test all strategies on SPDR
strategies = [
    ('sma_crossover', {'fast_period': 20, 'slow_period': 50}),
    ('rsi_mean_reversion', {'oversold': 30, 'overbought': 70}),
    ('macd_momentum', {}),
    ('bollinger_bands', {}),
    ('dual_momentum', {})
]

strategy_results = []

for strategy_name, params in strategies:
    print(f"\nTesting {strategy_name}...")
    
    bt_test = CryptoBacktester(asset_symbol="SPDR", initial_capital=10000)
    bt_test.data = bt_spdr.data.copy()  # Reuse data
    
    metrics = bt_test.run_strategy(strategy_name, allow_short=True, **params)
    
    strategy_results.append({
        'Strategy': strategy_name,
        'Return %': metrics['total_return'],
        'Win Rate %': metrics['win_rate'],
        'Sharpe': metrics['sharpe_ratio'],
        'Trades': metrics['total_trades']
    })

strategy_df = pd.DataFrame(strategy_results).sort_values('Return %', ascending=False)

print("\n" + "="*70)
print("STRATEGY COMPARISON ON SPDR")
print("="*70)
display(strategy_df)
```
